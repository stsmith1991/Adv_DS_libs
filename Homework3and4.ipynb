# Тема "Визуализация данных в Matplotlib"

```
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```
x=[1, 2, 3, 4, 5, 6, 7]
y=[3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
if len(x)==len(y):
  plt.plot(x, y)
  plt.show()
```

Задание 2.

```
t=np.linspace( 1, 10, 51, endpoint=True)
f=np.array(np.cos(t))
plt.plot(t, f, 'g')
plt.title('График f(t)',
          fontsize=18,
          fontweight="bold",
          family="arial",
          )
plt.xlabel('Значения t',
           fontsize=15)
plt.ylabel('Значения f',
           fontsize=15)
plt.axis([ 0.5, 9.5, -2.5, 2.5])
plt.gcf().set_facecolor("gray")
```

Задание 3.

```
#Задаю числовые данные графиков
x = np.linspace(-3, 3, 51)
y = x ** 2
y2 = 2*x+0.5
y3 = -3*x+1.5
y4 = np.sin(x)
 
#Задаю характеристики всего рисунка
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
 
#Задаю характеристики отдельных графиков
ax[0, 1].set_title("График 1")
ax[0, 1].axis(xmin=-5, xmax=5)
 
ax[1, 1].set_title("График 2")
 
ax[0, 0].set_title("График 3")
 
ax[1, 0].set_title("График 4")
 
#Отдаю команду на вывод фигуры с axis-ами
ax[0, 1].plot(x, y)
ax[1, 1].plot(x, y2)
ax[0, 0].plot(x, y3)
ax[1, 0].plot(x, y4)
```
